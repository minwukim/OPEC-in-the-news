### First, we import the pacakges and the news data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('news_data.csv')
df.head()
```

### Then, we check all the newspaper with the earliest year-month and the latest year-month

```python
# Convert the 'date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract year and month
df['year_month'] = df['Date'].dt.to_period('M')

# Group by 'Publication Title' and get min and max of 'year_month'
result = df.groupby('Publication Title')['year_month'].agg(['min', 'max']).reset_index()

print(result)
```

### Then, Filter out the publications whose latest year-month is 2023-10

```python
# Filter the rows where 'max' is '2023-10'
filtered_result = result[result['max'] == '2023-10']
print(filtered_result)
print("The maximum min value among the filtered results is:", filtered_result['min'].max())
```

### As the earliest year-month of Gulf News and Saudl Gazette are too late, remove these two as well. 

```python
# Filter the rows where 'max' is '2023-10' and exclude specified publication titles
filtered_result = result[
    (result['max'] == '2023-10') & 
    (result["Publication Title"] != "Saudi Gazette") & 
    (result["Publication Title"] != "Gulf News")
]
print(filtered_result)
print("The maximum min value among the filtered results is:", filtered_result['min'].max())
```

```python
publications = filtered_result["Publication Title"].tolist()
print(publications)
```

### Compute the newspaper index following the instruction of the paper.
1. Compute the standard deviation for each newspaper
2. devide each count by the respective standard deviation

```python
# Filter the df
filtered_df = df[df['Publication Title'].isin(publications)]

# Group and count
grouped = filtered_df.groupby(['Publication Title', 'year_month']).size().reset_index(name='count')

# Pivot
pivot_table = grouped.pivot(index='Publication Title', columns='year_month', values='count').fillna(0)

# Ensure all months are present
all_months = pd.period_range('1991-04', '2023-10', freq='M')
pivot_table = pivot_table.reindex(columns=all_months, fill_value=0)

# Reset index
newspaper_index = pivot_table.reset_index()

newspaper_index

std_devs = newspaper_index.iloc[:, 1:].std(axis=1) 

# 2. Divide each count by the respective standard deviation
normalized_counts = newspaper_index.iloc[:, 1:].divide(std_devs, axis=0)

# Update the counts in the dataframe
newspaper_index.iloc[:, 1:] = normalized_counts

newspaper_index
```

### Plot the data over time

```python
plot_data = newspaper_index.set_index('Publication Title').transpose()

# Convert Period objects to strings for plotting
plot_data.index = plot_data.index.astype(str)

plt.figure(figsize=(15, 7))

# Plotting each publication's data
for publication in publications:
    plt.plot(plot_data.index, plot_data[publication], label=publication)

# Create a list of month labels that only includes the beginning of each year
month_labels = [month if '-01' in month else '' for month in plot_data.index]

plt.title("Counts of Publications by Month")
plt.xlabel("Month")
plt.ylabel("Count")
plt.legend()
plt.xticks(plot_data.index, month_labels, rotation=45)  # Rotate x-axis labels for better visibility and only show January of each year
# plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Compute the index as instructed
1. Compute the average normalized count of rall the publication titles for each month (Z_t)
2. Compute the average of Z_t, denoted as M
3. Compute the final values as Z_t * 100 / M


```python
# 1. Compute average count for all publication titles for each month
Z_t = newspaper_index.iloc[:, 1:].mean()

# 2. Compute the average of Z_t
M = Z_t.mean()

# 3. Multiply Z_t by 100/M
final_values = Z_t * (100/M)

# 4. Insert the new row
index_row = ['index'] + final_values.tolist()
newspaper_index.loc[len(newspaper_index)] = index_row

newspaper_index
```

#### Plot the index over time

```python
# Extract the last row values excluding "Publication Title"
data_to_plot = newspaper_index.iloc[-1, 1:].values

# Extract the column names (which are the months) for the x-axis and convert them to string
months = [str(month) for month in newspaper_index.columns[1:]]

# Create a list of month labels that only includes the beginning of each year
month_labels = [month if '-01' in month else '' for month in months]

# Plot
plt.figure(figsize=(15, 7))
plt.plot(months, data_to_plot)
plt.title("Values for 'index' Publication Title")
plt.xlabel("Month")
plt.ylabel("Value")
plt.xticks(months, month_labels, rotation=45)
plt.tight_layout()
plt.show()
```

#### Plot the correlation matrix among different newspaper

```python
# Calculate the correlation matrix excluding the last row
corr_matrix = newspaper_index.iloc[:-1, 1:].transpose().corr()

# Get the publication titles excluding the last one
publication_titles = newspaper_index["Publication Title"].iloc[:-1]

# Plotting the heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=True, square=True, fmt=".2f",
            xticklabels=publication_titles, yticklabels=publication_titles)
plt.title("Correlation Matrix of Publication Titles")
plt.xticks(rotation=40)
plt.yticks(rotation=0)
plt.show()
```

### Load Google Search Volume Index and compute the correlation with the newspaper index

```python
gsv = pd.read_csv('GSV_data.csv')
gsv.head()
```

```python
# 1. Extract the row from newspaper_index where "Publication Title" is "index"
newspaper_data = newspaper_index[newspaper_index['Publication Title'] == 'index'].T.drop('Publication Title').squeeze()

# Ensure it's a PeriodIndex for consistency in plotting
newspaper_data.index = pd.PeriodIndex(newspaper_data.index)

# Filter data from 2004-01 to 2023-10 for newspaper_data (in case there's extra data outside this range)
newspaper_data = newspaper_data['2004-01':'2023-10']

# 2. Extract data from gsv dataframe
# Convert the 'month' column to Period type for consistency
gsv['Month'] = pd.PeriodIndex(gsv['Month'], freq='M')

# Set 'month' as the index for gsv dataframe
gsv_data = gsv.set_index('Month')['index']

# 3. Plot the data
plt.figure(figsize=(14,7))
plt.plot(newspaper_data.index.to_timestamp(), newspaper_data.values, label='Newspaper Index', marker='o')
plt.plot(gsv_data.index.to_timestamp(), gsv_data.values, label='GSV Index', linestyle='--')
plt.title('Newspaper Index vs GSV Index')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Calculate means
mean_x = newspaper_data.mean()
mean_y = gsv_data.mean()

# Calculate the numerator and the denominators
numerator = ((newspaper_data - mean_x) * (gsv_data - mean_y)).sum()
denom_x = ((newspaper_data - mean_x)**2).sum() ** 0.5
denom_y = ((gsv_data - mean_y)**2).sum() ** 0.5

# Calculate the correlation
correlation = numerator / (denom_x * denom_y)

print(f"Correlation between Newspaper Index and GSV Index: {correlation:.4f}")
```
